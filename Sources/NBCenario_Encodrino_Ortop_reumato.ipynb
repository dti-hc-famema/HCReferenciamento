```python
import numpy  as np
import pandas as pd
import matplotlib.pyplot as pl
import seaborn as sb
import random, decimal

%matplotlib inline
```

```python
df = pd.read_csv('DS-Encodrino_Ortop_reumato.csv',encoding="ISO-8859-1")
```

<h2>Plotando o dataset puro </h2>



```python
sb.pairplot(df, hue='EQUIPE')
```

```python
df.head()
```

```python
X = np.array(df.drop('EQUIPE',axis = 1))
```

## K-means com 7 cluster

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=7, random_state=0)
```

```python
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
df['K-classes'] = kmeans.labels_
```

<h2>Plotando o resultado da clusterização</h2>

```python
sb.pairplot(df,hue='K-classes')
```
